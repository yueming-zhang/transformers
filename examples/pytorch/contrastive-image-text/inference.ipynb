```python
import EmbedStore
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


store = EmbedStore.EmbedStore('train2017')
```

```python
root_path = '/mnt/e/ML.Data/coco_extracted/train2017/'

fig = plt.figure(figsize=(20, 20))
columns = 5
rows = 1
ret = store.search_faiss_index(['two dogs playing in the snow'], top_k=columns*rows)
print (ret)
for i in range(0, columns*rows):
    img_path = root_path + ret[0][i]
    img = mpimg.imread(img_path)
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(img)
plt.show()
```
