```python
from EmbedStore import EmbedStore
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

pic_folder = "train2017"
store = EmbedStore(pic_folder, 'checkpoint-13500')
```

```python
root_path = f'/mnt/e/ML.Data/coco_extracted/{pic_folder}/'

fig = plt.figure(figsize=(20, 20))
columns = 5
rows = 1
ret = store.search_faiss_index(['a dog using a red cellphone'], top_k=columns*rows)
print (ret)
for i in range(0, columns*rows):
    img_path = root_path + ret[0][i]
    img = mpimg.imread(img_path)
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(img)
plt.show()
```
